```python
import datetime

print("Welcome to the Citizen Information System!")

while True:
    choice = int(input('''Kindly choose from the following options by entering the number next to them:
                    1 : General Information
                    2 : Personal Information
                    3 : Address Details
                    4 : Education Details '''))
    if choice == 1:
        print("Please enter the following General Information")
        first_name = input("First Name: ")
        middle_name = input("Middle Name: ")
        last_name = input("Last Name: ")
        aadhar_no = input("Aadhar Number: ")
        aadhar_name = input("Name (as in aadhar card): ")
        pan_no = input("PAN Number: ")
        pan_name = input("Name (as in PAN card): ")
        n=input("do you want to add mmore information: Y/N ")
        if n== 'Y':
            continue
        elif n== 'N':
            break
        else:
            print("INVALID INPUT")
    elif choice == 2:
        print("Please enter your the following Personal Information")
        contact_no = input("Contact Number: ")
        email_id = input("Email ID: ")
        dob_str = input("Date of Birth (yyyy/mm/dd): ")
        dob = datetime.datetime.strptime(dob_str, "%Y/%m/%d")
        father_name = input("Father's Name: ")
        nationality = input("Nationality: ")
        blood_group = input("Blood Group: ")
        n=input("do you want to add mmore information: Y/N ")
        if n== 'Y':
            continue
        elif n== 'N':
            break
        else:
            print("INVALID INPUT")
    elif choice == 3:
        print("Please enter your following Address Details")
        house_no = input("House/Flat Number: ")
        locality = input("Street / Locality: ")
        town = input("Town: ")
        district = input("District: ")
        state = input("State: ")
        pincode = input("Pincode: ")
        post_office = input("Post Office: ")
        n=input("do you want to add mmore information: Y/N ")
        if n== 'Y':
            continue
        elif n== 'N':
            break
        else:
            print("INVALID INPUT")
    elif choice == 4:
        print("Please enter your following Education Details")
        high_school_name = input("High School Name: ")
        high_school_board = input("High School Board: ")
        inter_school_name = input("Senior Secondary School Name: ")
        inter_school_board = input("Senior Secondary School Board: ")
        degree = input("Latest Degree: ")
        n=input("do you want to add more information: Y/N ")
        if n== 'Y':
            continue
        elif n== 'N':
            break
        else:
            print("INVALID INPUT")
    else:
        print("Error: Invalid Choice. Please Try Again")

display = int(input("Enter the above numbers to display the corresponding details: "))
if display == 1:
    print("First Name: ", first_name)
    print("Middle Name: ", middle_name)
    print("Last Name: ", last_name)
    print("Aadhar Number: ", aadhar_no)
    print("Aadhar Name: ", aadhar_name)
    print("PAN Number: ", pan_no)
    print("PAN Name: ", pan_name)
elif display == 2:
    print("Contact Number: ", contact_no)
    print("Email ID: ", email_id)
    print("Date of Birth: ", dob)
    print("Father's Name: ", father_name)
    print("Nationality: ", nationality)
    print("Blood Group: ", blood_group)
elif display == 3:
    print("House/Flat Number: ", house_no)
    print("Street / Locality: ", locality)
    print("Town: ", town)
    print("District: ", district)
    print("State: ", state)
    print("Pincode: ", pincode)
    print("Post Office: ", post_office)
elif display == 4:
    print("High School Name: ", high_school_name)
    print("High School Board: ", high_school_board)
    print("Senior Secondary School Name: ", inter_school_name)
    print("Senior Secondary School Board: ", inter_school_board)
    print("Latest Degree: ", degree)
else:
    print("Error: Invalid Choice.")
```
